## match-case Statement

The match case statement, introduced in Python 3.10, is used for structural pattern matching. It evaluates a subject expression against a series of patterns and executes the code block for the first match. It is a more powerful and readable alternative to long if-elif-else chains when dealing with multiple conditions or complex data structures.

**Syntax**:

```python
match subject_expression:
    case pattern1:
        # Code block 1
    case pattern2:
        # Code block 2
    case _:
        # Default code block (wildcard)
```

- `match` is the keyword that initiates the statement.
- `subject_expression` is the value you are comparing.
- `case` defines a pattern to match against the subject.
- `_ `(underscore) acts as a wildcard and matches any subject if no other case matches, serving as a default.

### Matching Constants

````python
def http_error(status):
    match status:
        case 400:
            print("Bad request")
        case 401:
            print("Unauthorized")
        case 403:
            print("Forbidden")
        case 404:
            return "Not found"
        case _:
            print("Not a valid response code")
````

````python
http_error(405)
````

````python
http_error(403)
````

### Combining Cases (OR operator)

````python
def access(user):
    match user:
        case "admin" | "manager":
            return "Full access"
        case "Guest":
            return "Limited access"
        case _:
            return "No access"
````

````python
print(access("manager"))
````

````python
print(access("engineer"))
````

### Matching Sequences (Lists and Tuples)

````python
def process_data(data):
    match data:
        case [x, y]:
            return f"Two elements: {x} and {y}"
        case [x, y, z]:
            return f"Three elements: {x}, {y}, and {z}"
        case _:
            return "Unknown data format or length"
````

````python
print(process_data([1, 2]))
````

````python
print(process_data((1, 2, 3))) # Tuples and lists are treated the same
````

### Guards (Adding an if condition)

````python
def grade_student(score):
    match score:
        case score if score >= 90:
            return "A"
        case score if score >= 80:
            return "B"
        case _:
            return "F"
````

````python
print(grade_student(95))
````

### Matching Class Instances

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

def where_is(point):
    match point:
        case Point(x=0, y=0):
            print("Origin")
        case Point(x=0, y=y):
            print(f"Y={y}")
        case Point(x=x, y=0):
            print(f"X={x}")
        case Point():
            print("Somewhere else")
````

````python
where_is(Point(0, 5))
````

````python
where_is(Point(5, 5))
````

### Matching Class Instances with positional matching

````python
class Point:
    __match_args__ = ('x', 'y') # Defines positional matching
    def __init__(self, x, y):
        self.x = x
        self.y = y

def where_is(point):
    match point:
        case Point(x=0, y=0):
            print("Origin")
        case Point(x=0, y=y):
            print(f"Y={y}")
        case Point(x, 0): # Positional matching works due to __match_args__
            print(f"X={x}")
        case Point():
            print("Somewhere else")

where_is(Point(5, 0))
````
