# Functions and Variables

## Basic Output with `print()` built-in function

- The `print()` function displays output to the console.
- Each call of `print()` function prints on a new line by default.

```python
print("Hello")
print("Hello, world")
```

## Taking User Input

- The `input()` function reads user input as a string.
- The text inside parentheses is a prompt shown to the user.

```python
name = input("What's your name? ")
```

## Printing with Multiple Arguments

- `print()` can take multiple arguments.
- By default, Python inserts a space ("` `") between them.

```python
print("Hello,", name)
```

## String Concatenation

- Uses `+` to join strings.
- Works only when both operands are strings.

```python
print("Hello, " + name)
```

```python
print("Age: " + 25)
```

### The fix

```python
print("Age: " + str(25))
```

## f-Strings

- Modern and clean way to format strings.
- Expressions can be embedded inside `{}`.

```python
print(f"Hello, {name}")
```

## Controlling Line End (`end` parameter)

- Default : `end="\n"` (new line).
- Changing `end=" "` keeps output on the same line.

```python
print("Hello,", end=" ")
print(name)
```

## Custom Separator (`sep` parameter)

- `sep` defines what goes between arguments.
- Default : `sep=" "` (space).

```python
print("Hello,", name, sep="???")
```

## String Methods

### `strip()`
Removes whitespace from both ends.

```python
"     Ravi   Chandra      ".strip()
```

### `title()`

Capitalizes first letter of each word.

```python
"ravi chandra".title()
```

### `capitalize()`
- Only first letter of the entire string is capitalized.
- Not ideal for full names → use `title()` instead

```python
"ravi chandra".capitalize()
```

```python
name = input("What's your name? ")
name = name.strip()
name = name.title()
print(f"Hello, {name}")
```

### Splitting Strings & Extracting Parts

- `split(" ")` → splits the string using space as delimiter.
- Returns a list of words.

```python
"Ravi Chandra".split(" ")
```

```python
first, last = name.split(" ")
```

```python
print(f"Hello, {first}")
```

`first, last = name.split(" ")` will crash if input is not exactly 2 words.

```python
name = "ravi chandra k"
first, last = name.split(" ")
```

```python
name = "ravi chandra k"
name_sp = name.split()
first = name_sp[0].strip().title()
print(f"Hello, {first}")
```

### Flexible Unpacking

`first, *rest = name.split()`
- `first` → first name
- `rest` → remaining parts (list)

```python
first, *rest = name.split()
first
```

### Limit Split

```python
first, last = name.split(" ", 1)
```

```python
first
```

```python
name = input("Enter your name: ").strip().title()
first = name.split()[0]

print(f"Hello, {first}")
```
