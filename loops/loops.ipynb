# Loops

## `while` loop

A `while` loop in Python repeatedly executes a block of code as long as a specified condition remains `True`. This loop type is typically used when the number of iterations is not known in advance, such as when processing user input or waiting for a specific event.

**Syntax**:

```python
# Initialization (optional but often necessary)
count = 0 
while condition:
    # Code block to be executed repeatedly
    print(count)
    # Update the variable used in the condition
    count = count + 1 
```

The indented statements form the body of the loop. It is crucial to ensure that the condition will eventually become `False`, otherwise, the loop will run indefinitely (an infinite loop).

- **`while` keyword**: Initiates the loop.
- **Condition**: A Boolean expression evaluated before each iteration. If `True`, the loop body executes; if `False`, the loop terminates. Non-zero values are interpreted as `True`, and `0` or `None` as `False`.
- **Indented body**: One or more statements executed per iteration.
- **Variable update**: Code within the loop body that modifies the variable(s) involved in the condition, moving the loop closer to termination.

````python
count = 0 
while count < 5:
    # Code block to be executed repeatedly
    print(count)
    # Update the variable used in the condition
    count = count + 1 
````

## `while` Loop Control Statements

You can alter the flow of a `while` loop using control statements.

### `break`

Immediately terminates the entire loop, regardless of the condition's status.

````python
i = 1
while i < 30:
    print(i)
    if i == 3:
        break # Loop stops here
    i += 1
````

### `continue`

Skips the current iteration and returns control to the top of the loop, where the condition is re-evaluated.

````python
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue # Skips the print statement for i=3
    print(i)
````

### `else` clause 
An optional block that executes only after the loop condition becomes `False` naturally (i.e., not terminated by a break statement).

````python
i = 1
while i < 6:
    print(i)
    i += 1
else:
    print("i is no longer less than 6") 
````

### Validating Input

````python
while True:
    n = int(input("What's n? "))
    if n < 0:
        continue
    else:
        break
````

````python
while True:
    n = int(input("What's n? "))
    if n > 0:
        break
````

## `for` loop

A `for` loop in Python is used for iterating over a sequence (such as a list, tuple, string, or range) or other iterable objects. It repeatedly executes a block of code once for each item in the sequence, making the code concise and readable.

**Syntax**:

```python
for variable in sequence:
    # Code block to be executed repeatedly
    # This block must be indented
```

- `for` and `in` are keywords.
- `variable` is a temporary, arbitrary name that holds the current item's value in each iteration.
- `sequence` is the iterable object you want to loop over.
- The indented block of code (the loop body) is executed for each item.

### Iterating over a list

````python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
````

### Iterating through a string

````python
for char in "Python":
    print(char)
````

### Using `range()` for a sequence of numbers

The `range()` function generates a sequence of numbers and is ideal when you need to repeat a loop a specific number of times.

````python
for i in range(3):
    print(i)
````

````python
for _ in range(3):
    print("Hello")
````

````python
fruits = ["apple", "banana", "cherry"]
for i in range(len(fruits)):
    print(i, fruits[i])
````

### Iterating over a dictionary

By default, looping over a dictionary iterates through its keys. You can use the `.items()` method to access both keys and values.

````python
fruits = {'apple': 'red', "banana": 'yellow', "kiwi": "green"}
````

````python
fruits["apple"]
````

````python
fruits["kiwi"]
````

````python
# By default, looping over a dictionary iterates through its keys
for fruit in fruits:
    print(fruit)
````

````python
for fruit in fruits:
    print(fruits[fruit])
````

````python
for item in fruits.items():
    print(item)
````

````python
for key, value in fruits.items():
    print(f"Fruit: {key} \tColour: {value}")
````

### Iterate over a list of dictionaries

````python
data_list = [
    {'name': 'Alice', 'age': 30},
    {'name': 'Bob', 'age': 25},
    {'name': 'Charlie', 'age': 35}
]
````

````python
for dictionary in data_list:
    print(dictionary)
````

````python
for dictionary in data_list:
    print(f"{dictionary["name"]} \t {dictionary["age"]}")
````

## Nested Loops

````python
height = 3
width = 8
````

````python
for i in range(height):
    for j in range(width):
        print("#", end="")
    print()
````
