# Installing Python Packages

## Dealing with Multiple Python Packages

When you've got multiple Python versions installed, so running a script with a specific Python version is all about explicitly choosing the interpreter.

#### Run using full path
```bash
C:\Python314\python.exe your_script.py
```

#### Use py launcher
```bash
py -3.14 your_script.py
```

#### Install a package with specific Python
```bash
C:\Python314\python.exe -m pip install requests
```

#### Inside VS Code
- Press `Ctrl + Shift + P`.
- Select: **Python**: **Select Interpreter**
- Choose: `Python 3.14 (C:\Python314\python.exe)`

#### Inside Jupyter
Switch kernel → Python 3.14
